function to clean up unkown data by looking at trees of same adress and then picking the value that occurs most for the trees at this adress

```python
def cleanup(df,column_name):
    j=0
    longest=0
    for i in range(100000):
        spc_latin_index_list=[]
        spc=[]
        val=df.loc[i,column_name]
        if val=='unknown':
            j=j+1
            address=df.loc[i,'address']
            #list of all indexes with same adress
            spc_latin_index_list=df.index[df['address'] == address].tolist()
            #print(spc_latin_index_list[0])
            #print(df.loc[spc_latin_index_list[0],column_name])
            #print(set(spc_latin_index_list))
            for k in spc_latin_index_list:
                if df.loc[k,column_name] != 'unknown':
                    spc.append(df.loc[k,column_name])          
        ser=pd.Series(spc)
        if len(ser)>1:
            value=ser.value_counts().idxmax()
            if value != 'unkown':
                df.loc[i,column_name]=value
                print(value)
                print(i)
    print(j)
    print(longest)
```
